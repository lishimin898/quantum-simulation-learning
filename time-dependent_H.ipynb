```python
from qutip import *
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N=10
a=destroy(N)
vac=basis(N,0)
t=np.linspace(0,1000,10000)
```

```python
H0 = a.dag()*a + 1/2.0
H1 = a.dag()*a.dag()+a*a
def H1_coeff(t,A,w):
    return A*np.cos(w*t)
H=[H0,[H1,H1_coeff(t,0.01,2)]]
psi0=vac
output=mesolve(H,psi0,t,c_ops=[],e_ops=[a.dag()*a])
```

```python
plt.figure(figsize=(16,9))
plt.plot(t,output.expect[0])
```
